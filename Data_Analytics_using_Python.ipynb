## Lambda Functions

lambda arguements: expression (lambdas use 1 or more arguements but ONLY 1 expression)

```python
# Basic lambda example. 1 arguement x
example = lambda f: f - 2

example(10)
```

```python
# Basic lambda example: multiple arguments. v, w, x, y, z
example_2 = lambda v, w, x, y, z: x * 2 + y - z +(v*6)-(v*w+3)

example_2(4, 6, 5, 10, 15)
```

```python
# Create a list.
sequence = [10, 2, 8, 7, 5, 4, 3, 11, 0, 1]

# Lambda and filter function. This filters the list called sequence for values greater than 6
filter_answer = filter (lambda x: x > 6, sequence)

# Call the output to be a list of the values over 6
list(filter_answer)
```

```python
# Lambda and the map() function
sequence_2 = [10, 2, 8, 7, 5, 4, 3, 11, 0, 1, 14, 67]

# Use lambda and map to square each value un the sequence. Use the arguement x, and x*x is x squared
squared = map(lambda x:x*x, sequence_2)

# Output the results as a list
list(squared)
```

```python
# Using lambda to combine values like first and last names. use the .strip() function to remove any invisible blanks
# title() changes lower case to upper
full_name = lambda fn, ln: fn.strip().title()+" "+ln.strip().title()

# Output returns name with caps and the spaces before and after names removed(stripped)!
full_name("      ben","      parkinson   ")
```
