# Q1 - How many films released each year from 2012 were watched on Netflix

```python
# Group Merged_1 by year and Netflix, then use agg function to sum up numbers, reset index to indicate 1 index per year
# Then return values from 2012 only

Q1 = Merged_1.groupby(['Year'])['Netflix'].agg('sum').reset_index()
Q1[Q1['Year'] >=2012]
```

# Q2 What is the average runtime of movies released each year?

```python
# Group by Year and runtime, then find average, then reset the index
Q2 = Merged_1.groupby(['Year'])['Runtime'].agg('mean').reset_index()

# Return values from 2012 onwards
Q2[Q2['Year'] >=2012]
```

# Q3 What are the best and worst reviews recieved on rotten tomatoes?

```python
# Group by year, and define min for rotten tomatoes scores. Reset the index and return values from 2012 onwards

Q3_min = Merged_1.groupby(['Year'])['Rotten Tomatoes'].agg('min').reset_index()
Q3_min[Q3_min['Year'] >=2012]
```

```python
# Group by year, and define max for rotten tomatoes scores. Reset the index and return values from 2012 onwards

Q3_max= Merged_1.groupby(['Year'])['Rotten Tomatoes'].agg('max').reset_index()
Q3_max[Q3_max['Year'] >=2012]
```

```python
# Or return both in same DataFrame

Q3_RT= Merged_1.groupby(['Year'])['Rotten Tomatoes'].agg(['min','max']).reset_index()
Q3_RT[Q3_RT['Year'] >=2012]
```
